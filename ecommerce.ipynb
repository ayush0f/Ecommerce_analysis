```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default= "plotly_white"
```

```python
data=pd.read_csv("/content/Sample - Superstore.csv", encoding='latin1')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

#converting date **coloumns**

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data['order month']= data['Order Date'].dt.month
data['order year']= data['Order Date'].dt.year
data['order day of week']=data['Order Date'].dt.dayofweek
```

```python
data.head()
```

# Monthly Sales **Analysis**

```python
sales_by_month=data.groupby('order month')['Sales'].sum().reset_index()
fig=px.line(sales_by_month, x='order month', y='Sales', title='Monthly Sales Analysis')
fig.show()
```

```python
data.head()
```

# SALES BY **CATEGORY**

```python
sales_by_category=data.groupby('Category')['Sales'].sum().reset_index()
fig=px.pie(sales_by_category, values='Sales', names='Category', title='Sales by Category',hole=0.5)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title='sales by category',uniformtext_minsize=12, uniformtext_mode='hide')
fig.show()
```

# **Sales analysis by sub cat**

```python
data.head()
```

```python
sales_by_sub_category=data.groupby('Sub-Category')['Sales'].sum().reset_index()
fig=px.bar(sales_by_sub_category, x='Sub-Category', y='Sales', title='Sales Analysis by Sub-Category')
fig.show()
```

# ***Mothly Profit Analysis***

```python
data.head()
```

```python
profit_by_month=data.groupby('order month')['Profit'].sum().reset_index()
fig=px.line(profit_by_month, x='order month', y='Profit', title='Monthly Profit Analysis')
fig.show()
```

# ***Profit by category***

```python
profit_by_category=data.groupby('Category')['Profit'].sum().reset_index()
fig=px.pie(profit_by_category, values='Profit', names='Category', title='Profit by Category',hole=0.5)
fig.show()
```

# ***SALES BY CUSTOMER SEGMENT***

```python
sales_profit_by_segment=data.groupby('Segment').agg({'Sales':'sum', 'Profit':'sum'}).reset_index()
fig=px.bar(sales_profit_by_segment, x='Segment', y=['Sales', 'Profit'], barmode='group', title='Sales and Profit by Customer Segment')
fig.show()
```

# ***Sales and profit Ratio***

```python
sales_profit_by_segment=data.groupby('Segment').agg({'Sales':'sum', 'Profit':'sum'}).reset_index()
sales_and_profit=sales_profit_by_segment['Profit']/sales_profit_by_segment['Sales']
sales_profit_by_segment['Sales and Profit Ratio']=sales_and_profit
display(sales_profit_by_segment)
fig=px.bar(sales_profit_by_segment, x='Segment', y='Sales and Profit Ratio')
fig.show()
```
